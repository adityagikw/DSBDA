```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset/titanic.csv')
df
```

```python
plt.figure(figsize=(10,8)) # plt.figure(): Creates a new figure window with a custom size (10 inches wide, 8 inches tall)

sns.histplot(data=df,x='fare',hue="survived",palette="Set1",kde=True)
# sns.histplot(): Creates a histogram plot. data=df: Uses the Titanic DataFrame. x='fare': Sets the x-axis to represent the fare paid by passengers. hue="survived": Colors the bars based on survival (0 = no, 1 = yes). palette="Set1": Uses a predefined color palette. kde=True: Adds a KDE (Kernel Density Estimate) line to show the smoothed distribution curve.

plt.title("histogram for titanic dataset") # title(): Sets the title of the chart.
plt.ylabel('passanger count') # ylabel(): Labels the y-axis as "passenger count".
plt.legend(title="survived",labels=['no','yes']) # legend(): Adds a legend to explain the color mapping.
plt.tight_layout() # This function adjusts the padding between and around subplots for a cleaner layout.
plt.show() # show(): Displays the final plot.
```
