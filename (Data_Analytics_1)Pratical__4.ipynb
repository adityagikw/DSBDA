Create a Linear Regression Model using Python/R to predict home prices using Boston Housing
Dataset (https://www.kaggle.com/c/boston-housing). The Boston Housing dataset contains
information about various houses in Boston through different parameters. There are 506 samples
and 14 feature variables in this dataset.
The objective is to predict the value of prices of the house using the given features.

```python
# x is the feature matrix, where all columns except medv (Median value of homes) are retained for training the model. y is the target variable, which is the medv column (house prices). x will contain the 13 features (e.g., crime rate, average number of rooms, etc.). y will be the house price (medv).

x = df.drop('medv',axis=1)
y = df['medv']
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=40)
print(x_train)

# train_test_split() splits the data into training and testing sets: x_train and y_train: Training data for features and target variable.
#                                                                    x_test and y_test: Test data for features and target variable.
#                                                                    test_size=0.2: 20% of the data will be used for testing, and 80% for training.
#                                                                    random_state=40: Ensures reproducibility by controlling the random splitting.
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)

# LinearRegression(): Initializes the linear regression model. model.fit(): Trains the model on the training data (x_train, y_train), allowing the model to learn the relationship between features and the target variable (house prices).
```

```python
y_pred = model.predict(x_test)  # model.predict(x_test): The trained model is used to predict house prices (y_pred) on the test data (x_test).
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print(mse)
print(r2)

# mean_squared_error: Computes the Mean Squared Error (MSE), which represents the average of the squared differences between the actual and predicted values. A lower MSE indicates better performance. r2_score: Computes the R-squared value (R²), which measures how well the model explains the variance in the target variable (house prices). R² ranges from 0 to 1. A value closer to 1 indicates a better fit. The print() statements display the computed MSE and R² values.
```
